```python
import numpy as np
import matplotlib.pyplot as plt
from random import randint
```

```python
def cost(x,y,theta,m):
    ret=0
    for i in range(0,m):
        ret+=(sum(x[i]*theta)-y[i])**2
    return (ret)
```

```python
X=[1,4,5,6]
Y=[-1,2,3,4]
X=np.asarray(X)
Y=np.asarray(Y)
```

```python
X=np.asarray(X)
X=[np.ones(X.shape),X]
X=np.asarray(X)
```

```python
X=X.transpose()
```

```python
Y=Y.transpose()
```

```python
theta=np.zeros((X[0].shape))
```

```python
def Gradient_descent(X,Y,alpha,iterations,theta):
    m=len(Y)
    for i in range(0,iterations):
        grad=X.T@ ((X@theta)-Y)
        theta=theta-(alpha/m) * grad;
       # print('Grad:{}'.format(grad))
       # print('Iteration : {} , cost : {}'.format(i,cost(X,Y,theta,m)))
    
    plt.figure()
    plt.scatter(X[:,1],Y)
    temp=[min(X[:,1]),max(X[:,1])]
    plt.plot(temp, np.matmul(np.asarray([[1]*len(temp),temp]).transpose() ,theta.transpose()))
    plt.show()
    
    return theta
```

```python
theta=np.zeros((X[0].shape))
Gradient_descent(X,Y,0.01,5000,theta)
```

```python
noise=1000
y1=np.linspace(3,1000,150)
y1=[y + randint(0,noise)-noise   for y in y1]
```

```python
x1=[i for i in range(0,len(y1))]
x1=np.asarray(x1)
y1=np.asarray(y1)
plt.scatter(x1,y1,s=10)
plt.show()
```

```python
x1=np.asarray(x1)
x1=[np.ones(x1.shape),x1]
x1=np.asarray(x1)
x1=x1.transpose()
```

# Feature Scaling

```python
for i in range(1,len(x1[0])):
    x1[:,i] = (x1[:,i]-np.mean(x1[:,i]) )/np.std(x1[:,i])
```

```python
theta1=np.zeros(x1[0].shape)
theta1=Gradient_descent(x1,y1,0.01,2000,theta1)
```
