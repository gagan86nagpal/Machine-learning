```python
import numpy as np
import matplotlib.pyplot as plt
from random import randint
```

```python
x=[1,4,5,6]
y=[-1,2,3,4]
x=np.asarray(x)
y=np.asarray(y)
```

```python
plt.figure()
plt.scatter(x,y)
plt.show()
```

```python
theta=np.asarray([0 ,0])
```

```python
X=[[1]*len(x) ,x]
X=np.asarray(X)
X=X.transpose()
```

```python
def cost(x,y,theta,m):
    ret=0
    for i in range(0,m):
        ret+=(sum(x[i]*theta)-y[i])**2
    return (ret)
```

```python
def grad(x,y,theta,m):
    grad=np.zeros(theta.shape)
    grad=np.matmul(x,theta) -y
    return grad
```

```python
alpha=0.09
m=len(y)
theta=np.asarray([0,0])
for i in range(0,1000):
    #print('Iteration : {} , cost : {}'.format(i,cost(X,y,theta,m)))
    theta=theta-(alpha/m)*  np.matmul(grad(X,y,theta,m) ,X)
plt.figure()
plt.scatter(x,y)
temp=[min(x),max(x)]
plt.plot(temp, np.matmul(np.asarray([[1]*len(temp),temp]).transpose() ,theta.transpose()))
plt.show()
theta
```

```python
Xt=X.transpose()
thetaOptimal= np.matmul(np.matmul (np.linalg.pinv(np.matmul(Xt,X)), Xt) ,y)
co=cost(X,y,thetaOptimal,m)
plt.figure()
plt.scatter(x,y)
temp=[min(x),max(x)]
plt.plot(temp, np.matmul(np.asarray([[1]*len(temp),temp]).transpose() ,thetaOptimal.transpose()))
plt.show()
```
